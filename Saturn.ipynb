```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def describe(a):
    '''Summarize the basic characteristics of an array.'''
    print(f"length is {len(a)}")
    print(f"size is {a.size}")
    print(f"shape is {a.shape}")
    print(f"number of dimensions is {a.ndim}")
    print(f"contents are...\n{a}")
    print(f"maximum value is {a.max()}")
```

```python
directory = ""
planet = np.load(directory + 'planet.npy')
```

```python
describe(planet)
```

```python
# display the 2D array as an image
plt.imshow(planet, cmap = 'gray')
# draw a colorbar
plt.colorbar(orientation = 'horizontal');
```

```python
plt.imshow(planet, cmap = 'cividis')
plt.colorbar(orientation = 'horizontal');
# horizontal line
row = 1100
plt.axhline(row, color = 'white');
```

```python
# extract row 1100 and every column
plt.plot(planet[row, :])
plt.xlabel('Column (pixel)')
plt.ylabel('Brightness along Row {}'.format(row));
```

*The middle is Saturn, the two on either side are the rings. The Cassini Division around 600 and 4000 and the Encke Gap near the very edge of the outer ring right beore it drops to zero.*
